```python
import numpy as np
import pickle
import gzip
import matplotlib.pyplot as plt
from sklearn import svm, metrics
import time
```

```python
def read_mnist(mnist_file):
    """
    Reads MNIST data.
    
    Parameters
    ----------
    mnist_file : string
        The name of the MNIST file (e.g., 'mnist.pkl.gz').
    
    Returns
    -------
    (train_X, train_Y, val_X, val_Y, test_X, test_Y) : tuple
        train_X : numpy array, shape (N=50000, d=784)
            Input vectors of the training set.
        train_Y: numpy array, shape (N=50000)
            Outputs of the training set.
        val_X : numpy array, shape (N=10000, d=784)
            Input vectors of the validation set.
        val_Y: numpy array, shape (N=10000)
            Outputs of the validation set.
        test_X : numpy array, shape (N=10000, d=784)
            Input vectors of the test set.
        test_Y: numpy array, shape (N=10000)
            Outputs of the test set.
    """
    f = gzip.open(mnist_file, 'rb')
    train_data, val_data, test_data = pickle.load(f, encoding='latin1')
    f.close()
    
    train_X, train_Y = train_data
    val_X, val_Y = val_data
    test_X, test_Y = test_data    
    
    return train_X, train_Y, val_X, val_Y, test_X, test_Y

train_X, train_Y, val_X, val_Y, test_X, test_Y = read_mnist('mnist.pkl.gz')
```

# LINEAR KERNEL


### C = [0.001, 0.01, 1, 10, 100, 1000]

```python
#TRAIN
# Mảng các giá trị của siêu tham số C
C  = [0.001, 0.01, 1, 10, 100, 1000]

#Khởi tạo mảng độ lỗi 
E_train_lin = []
E_val_lin = []

for i in C:
    print("Với C = ", i)
    clf = svm.SVC(C = i, kernel = 'linear') #Linear Kernel

    start_time = time.time()
    clf.fit(train_X, train_Y)
    print("Total time: %s seconds" % (time.time() - start_time))

    #Get the error value
    E_train_c = 1 - clf.score(train_X, train_Y)
    E_val_c = 1 - clf.score(val_X, val_Y)
    
    #Lưu độ lỗi vào mảng đã khởi tạo
    E_train_lin.append(E_train_c)
    E_val_lin.append(E_val_c)
    
    print("Độ lỗi trên tập training: ", E_train_c)
    print("Độ lỗi trên tập validation: ", E_val_c)
```

## Biểu đồ sự thay đổi độ chính xác trên tập train và tập test theo C

Tính giá trị Accuracy từ các mảng độ lỗi E_train_lin và E_val_lin.

```python
# Khai báo mảng độ chính xác trên tập training và tập validation
A_train_lin = []
A_val_lin = []

# Chuyển đổi độ lỗi thành độ chính xác (1-Error)
for i in E_train_lin:
    A_train_lin.append(1 - i)
for j in E_val_lin:
    A_val_lin.append(1 - j)
```

Vẽ biểu đồ độ chính xác trên tập Train và tập Validation

```python
import matplotlib.pyplot as plt

# Vẽ 3 biểu đồ cùng lúc với ax3, ax2, ax1
fig, (ax3, ax2, ax1) = plt.subplots(1,3, figsize = (25,10), dpi = 100)

# Biểu đồ sự thay đổi độ chính xác trên tập training và tập validation trên toàn mảng giá trị C[0.001-1000]
ax1.plot(C, A_train_lin, label = 'Accuracy on training')
ax1.plot(C, A_val_lin, label = 'Accuracy on validation')
ax1.set_xlabel('Cost')
ax1.set_ylabel('Accuracy')
ax1.set_title("Độ chính xác của hàm huấn luyện theo C[0.001, 0.01, 1, 10, 100, 1000]")
ax1.legend(loc = 'best')

# Biểu đồ sự thay đổi độ chính xác trên tập training và tập validation trên các C từ 0.001 0.01 và 1
ax2.plot([C[0], C[1], C[2]], [A_train_lin[0], A_train_lin[1], A_train_lin[2]], label = 'Accuracy on training')
ax2.plot([C[0], C[1], C[2]], [A_val_lin[0], A_val_lin[1], A_val_lin[2]], label = 'Accuracy on validation')
ax2.set_xlabel('Cost')
ax2.set_ylabel('Accuracy')
ax2.set_title("Độ chính xác của hàm huấn luyện theo C[0.001, 0.01, 1]")
ax2.legend(loc = 'best')

# Biểu đồ sự thay đổi độ chính xác trên tập training và tập validation trên các C từ 0.001 và 0.01
ax3.plot([C[0], C[1]], [A_train_lin[0], A_train_lin[1]], label = 'Accuracy on training')
ax3.plot([C[0], C[1]], [A_val_lin[0], A_val_lin[1]], label = 'Accuracy on validation')
ax3.set_xlabel('Cost')
ax3.set_ylabel('Accuracy')
ax3.set_title("Độ chính xác của hàm huấn luyện theo C[0.001, 0.01]")
ax3.legend(loc = 'best')

plt.show()
```

```python
# Train model với c tốt nhất vừa chọn được, sau đó tính độ lỗi trên tập testing.
from sklearn import svm
import time
c_best = 0.01
#TRAIN
clf = svm.SVC(C = c_best, kernel = 'linear') #Linear Kernel

start_time = time.time()
clf.fit(train_X, train_Y)
print("Total: %s seconds" % (time.time() - start_time)) 

#Get the error value
print("Độ lỗi trên tập test: ", 1 - clf.score(test_X, test_Y))
```

```python
# Lưu model đã train với c_best ra file 'mnist-linear-svm.joblib'
from joblib import dump, load
dump(clf, 'mnist-linear-svm.joblib') 
```

```python
# Cho thấy sự chính xác, độ chính xác của hàm dự đoán cuối cùng sau khi train model với c_best
#PREDICT
predict = clf.predict(test_X)

print("Classification report for classifier %s:\n%s\n"
      % (clf, metrics.classification_report(test_Y, predict)))
disp = metrics.plot_confusion_matrix(clf, test_X, test_Y)
disp.figure_.suptitle("Confusion Matrix")
#print("Confusion matrix:\n%s" % disp.confusion_matrix)

plt.show()
```
